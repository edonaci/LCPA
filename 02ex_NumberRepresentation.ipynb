```python

s=str(input("Quale numero in precisione singola vuoi che traduca?"))

if(len(s)!=32):
    print("Inserire 32 cifre!")
if(len(s)==32):
    def spfp(s):
        if s[0]=="0":
            segno="+"
        if s[0]=="1":
            segno="-"
        m=0
        e=-127

        exp=s[1:9]


        i=0
        while(i<len(exp)):
            if exp[i]=="1":
                e=e+2**(7-i)
            i=i+1

        mantissa=s[9:]

        j=0
        while(j<len(mantissa)):
            if mantissa[j]=="1":
                m=m+pow(2,-(j+1))
            j=j+1    

        numero=segno+str((1+m)*pow(2,e))
        return(numero)
    
    print("il numero in precisione singola è : ",spfp(s))
```

7\. Consider the integral of the semicircle of radius 1:
$$
I=\int_{-1}^{1} \sqrt(1-x^2) {\rm d}x
$$
which it's known to be $I=\frac{\pi}{2}=1.57079632679...$.
Alternatively we can use the Riemann definition of the integral:
$$
I=\lim_{N\to\infty} \sum_{k=1}^{N} h y_k 
$$

with $h=2/N$ the width of each of the $N$ slices the domain is divided into and where
$y_k$ is the value of the function at the $k-$th slice.

(a) Write a programe to compute the integral with $N=100$. How does the result compares to the true value?

(b) How much can $N$ be increased if the computation needs to be run in less than a second? What is the gain in running it for 1 minute? 


```python
import math
import timeit

N=int(input("In quanti intervalli vuoi dividere l'intervallo di integrazione?"))

def semicircle(x):
    y=math.sqrt(1-x*x)
    return y

integral=0.0
x=-1

while(x<=1.0):
    integral=integral+semicircle(x)*2/N
    x=x+2/N
     

err_rel=abs( ( (integral-math.pi/2)/(math.pi/2) ) )*100
print("Per N =",N,"l'errore percentuale è ",err_rel)


#il valore massimo di N per rimanere all'interno del secondo di esecuzione è 1 000 000
#l'errore percentuale in questo caso è 1.0397600784467428e-07
#Per un minuto di running,cioè circa 50 000 000 di iterazioni ho un errore percentuale di circa
#3.1921234190248934e-08 . Quindi il guadagno è di un fattore 10
```
